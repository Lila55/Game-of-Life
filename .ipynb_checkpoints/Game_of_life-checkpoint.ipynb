```python
import tkinter as tk
from PIL import ImageTk , Image
import numpy as np
import time

global grillei,grillef

#r√©alister la grille------------------------------------!
dim = 17
nbl= 20
nbc=20

#les Widgets --------------------------------------------!
fen = tk.Tk()
can = tk.Canvas(fen,width = nbc*(dim), heigh=nbl*(dim) ,bg="gray")

img = ImageTk.PhotoImage(Image.open('image/blanc.png'))  # PIL solution
imgn = ImageTk.PhotoImage(Image.open('image/noir.png'))  # PIL solution


global stop 
stop = False
global comm
comm = False


def nouvelle_grille():   
    global grillei,grillef,stop,comm
    grillei= np.zeros((nbl,nbc))
    grillef= np.zeros((nbl,nbc))
    stop = False
    comm = False
    
    for i in range(nbl):
        for j in range(nbc):
            centre=(i*dim+dim/2,j*dim+dim/2)
            can.create_image(centre,image=img)
    can.bind("<Button>",select_cellule)       


def select_cellule(event):
    global grillei
    x=event.x
    y=event.y
    i=x//dim
    j=y//dim
    centre=(i*dim+dim/2,j*dim+dim/2)
    can.create_image(centre,image=imgn)
    grillei[i,j]=1
    
    
def animation():
    global stop
  
    global grillei,grillef
    if (stop):
        stop=False
        return
    for i in range (nbl):
        for j in range(nbc):
            
            centre=(i*dim+dim/2,j*dim+dim/2)
            if (i==0 ):
                grillef[i][j]=0 
                can.create_image(centre,image=img)

            elif (i==nbl-1):
                grillef[i][j]=0
                can.create_image(centre,image=img)
           
            elif(j==0):
                grillef[i][j]=0
                can.create_image(centre,image=img)

            elif(j==nbc-1):
                grillef[i][j]=0
                can.create_image(centre,image=img)
            else :
                v=grillei[i-1,j-1]+grillei[i,j-1]+grillei[i+1,j-1]+grillei[i-1,j]+grillei[i+1,j]+grillei[i-1,j+1]+grillei[i,j+1]+ grillei[i+1,j+1]
                
                if v>3 or v<2:
                    grillef[i][j]=0
                                                                                                                                     
                    can.create_image(centre,image=img)
                if((v==2 ) and grillei[i,j]==1):
                    grillef[i][j]=1
                    can.create_image(centre,image=imgn)
                if(v==3):
                    grillef[i][j]=1
                    
                    can.create_image(centre,image=imgn)            
    if np.array_equal(grillei,grillef):
        print('Etat stable')
        return
    grillei = np.array(grillef, copy=True)
    can.after(1,animation)
   

def an():
    global comm 
    if(comm):
        return
    comm=True
    animation()

def arret():
    global stop 
    global comm
    stop=True
    comm = False
    
    
    
nouvelle_grille()


btn=tk.Button(fen , text="commencer",command=animation)
btna=tk.Button(fen , text="arreter",command=arret)
btnn=tk.Button(fen , text="Nouvelle grille",command=nouvelle_grille)

can.pack(side=tk.LEFT)
btnn.pack(padx=10,pady=10)
btn.pack(padx=30,pady=20)
btna.pack(padx=20,pady=30)

fen.mainloop()
```
